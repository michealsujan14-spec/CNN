```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

# Features:
# [square_feet, bedrooms]
x = np.array([
    [800, 2],
    [1200, 3],
    [1500, 3],
    [2000, 4],
    [2500, 4],
    [3000, 5]
])

# Prices in lakhs
y = np.array([30, 45, 55, 80, 95, 120])

# Split data
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Advanced Polynomial Regression Model
model =make_pipeline(
    PolynomialFeatures(degree=2),
    LinearRegression()
)

# Train model
model.fit(x_train, y_train)

# Prediction
predicted_price = model.predict([[1800, 3]])

print("Predicted Price:", predicted_price[0], "Lakhs")

# Visualization using square feet only
sqft = x[:, 0]

# Sort for smooth graph
sorted_indices = np.argsort(sqft)

sqft_sorted = sqft[sorted_indices]
y_sorted = y[sorted_indices]

# Predict for plotting
plot_x = np.column_stack((sqft_sorted, np.full(len(sqft_sorted), 3)))
plot_y = model.predict(plot_x)

# Plot
plt.figure(figsize=(8, 5))

plt.plot(sqft, y, label="Actual Prices" , marker= "o" )
plt.plot(sqft_sorted, plot_y, label="ML Prediction Line", marker= "o")

plt.xlabel("Square Feet")
plt.ylabel("Price (Lakhs)")
plt.title("Advanced House Price Prediction")

plt.legend()
plt.grid(True)

plt.show()
```
